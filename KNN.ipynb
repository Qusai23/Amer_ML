### (b.i) (2 points) 
Classification: Implement a K-Nearest Neighbor classifier (K-NN) using
the euclidean distance (l2-norm) as a distance measure to classify between the three classes.
Please implement K-NN and do not use available libraries. In the report, please show
your code.

```python
def KNN(testinstance,trainData,k):

    train = np.array(trainData)
    dev = np.array(devData)
    test = np.array(testData)

    trainClass = trainData['class'].to_list()
    testClass = trainData['class'].to_list()
    devClass = trainData['class'].to_list()

    distList = []
    for i in range(0,len(train)):
        dist = np.linalg.norm(testinstance-train[i][0:4])
        distList.append((dist,i))

    neighbors = (sorted(distList)[:k])
    #print(distList)
    #print(neighbors)

    classes = {"Iris-setosa":0,"Iris-versicolor":0,"Iris-virginica":0}

    for i in range(0,len(neighbors)):
        response = trainClass[neighbors[i][1]]


        if response in classes:
            classes[response] += 1
        else:
            classes[response] =1
    sortedVotes = sorted(classes.items(), key=itemgetter(1), reverse=True)
    result = sortedVotes[0][0]

    return (sortedVotes)
```

#### (b.ii) (1 point) Explore different values of K = 1, 3, 5, 7, . . . , 19. You will train one model for
each of the ten values of K using the train data and compute the classification accuracy (Acc)
of the model on the development set. Plot the Acc metric on the dev set against the different
values of K. Please report the best hyper-parameter K∗ based on the Acc metric. Please
implement this procedure from scratch and do not use available libraries.

```python
def score(k):
    dev = np.array(devData)
    AccList = []

    for i in range(0,len(dev)):
        test = dev[i][0:4]
        results = KNN(test,trainData,k)
        Acc = results[0][1]/k
        AccList += [Acc]
    #print(AccList)
    score = sum(AccList)/len(AccList)
    return score
```

```python
scoreList = []
kVals = [i for i in range(0,20) if i%2 !=0]
for k in kVals :
    scoreList += [score(k)]
#print(score)
print(scoreList)

#plot the ACC metric on dev vs k values

plt.plot(kVals,scoreList)
plt.title('Dev Acc vs kVals', fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.xlabel('k values',fontdict = {'fontsize' : 14, 'fontname':'Comic Sans MS'})
plt.ylabel('Accuracy 100%',fontdict = {'fontsize' : 14, 'fontname':'Comic Sans MS'})

plt.show

```
