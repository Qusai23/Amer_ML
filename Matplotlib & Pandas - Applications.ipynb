```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line Graphs

```python
gas = pd.read_csv('gas_prices.csv')
gas.head(4)
```

```python

plt.figure(figsize = (8,5))
plt.plot(gas.Year,gas.USA, 'b.-', label='USA')
plt.plot(gas.Year,gas.Canada,'r.-', label='CA')
plt.plot(gas['Year'],gas['South Korea'],'g.-', label='SK')
plt.title('Oil Prices for US & CA', fontdict = {'fontname':'Comic Sans MS','fontsize':15})
plt.xlabel('Year',fontdict = {'fontsize' : 14, 'fontname':'Comic Sans MS'})
plt.ylabel('$/gal',fontdict = {'fontsize' : 14, 'fontname':'Comic Sans MS'})



plt.legend()
plt.xticks(gas.Year[::3]) # Start from the beginning and end from the end but skip 3
# Add even more years --> xtick(gas.Year[::3].tolist()+[2011])

plt.show()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country], marker = '.')
        

plt.show()
    
```

### Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins) # color using color picker
plt.xticks(bins)
plt.ylabel('Num of Players')
plt.xlabel('Skill Level')
plt.title('Disp of player skills in FIFA 2018')
plt.show
```

Pie chart

```python
# % of people that use left and right foot in Pie Chart

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.pie([left, right], labels = ['left', 'right'], colors = ['r','y'], autopct='%.2f%%')





plt.show()
```

```python
fifa.Weight
#remove ibs from weights

#fifa.Weight = [x.strip('Ibs') if type(x)==str else x for x in fifa.Weight]
#fifa.Weight
# Convert to int
fifa.Weight = [int (x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
```

```python
# Count light players

plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 200)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
lables = ['light','light_medium','medium','medium_heavy','heavy']
#plt.pie(weights, labels = lables)

# Note very squashed --> setting the percentage distance 

#plt.pie(weights, labels = lables, pctdistance=0.8)

# Move them apart
explode = (.4,.1,0,0,.4)
plt.pie(weights, labels = lables, pctdistance=0.8, explode=explode)

plt.show
```

```python
# Compare teams using boxplot
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
boston = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels = [ 'FCB', 'RMA','REV']
plt.boxplot([barcelona,madrid,boston], labels = labels)
plt.title('Professional Soccer Team')
plt.ylabel('Fifa over all rating')
plt.show()
```
