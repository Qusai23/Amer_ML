```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
#plt.plot([1,2,3], [2,4,6])

x = [0,1,2,3,4]
y = [0,2,4,6,8]
#plt.plot(x,y)
plt.plot(x,y, label='2x', color='red', linewidth = 2, marker='.', markersize = 10, markeredgecolor='y',linestyle = '--')

# add a 2ed line

x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2,x2**2,label='x^2')


# add title 
plt.title('Our first graph!!', fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('x_axis',fontdict={'fontname':'Comic Sans MS', 'fontsize':15})
plt.ylabel('y_axis',fontdict={'fontname':'Comic Sans MS', 'fontsize':15})

#ticks mark
plt.xticks([0,1,2,3, 4])
plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()
# Error because ther is nothing to label go back to plot(x,y) and add label

# Manuplate the line in plot function




plt.show() # To remove the line of words
```

```python
# Resize a graph

plt.figure(figsize=(5,3), dpi = 110) # try to keep value higher so print smooth

#Show projection of graph
x1 = np.arange(0,4.5,0.5)
plt.plot(x1[:6],x1[:6]**2,'r', label='2x')
plt.plot(x1[5:],x1[5:]**2,'y--')
plt.legend()

# save  a graph 
plt.savefig('mygraph.png',dpi=300)
plt.show()




```

# Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels,values)
# Costimize bar
# barbars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')

#-OR-
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# Resize 
plt.figure(figsize = (6,4))



plt.show()
```
